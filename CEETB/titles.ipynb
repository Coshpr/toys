```python
import xml
file_pth = "./titles.xml"

import os
if not os.path.isfile(file_pth):
    print("File not exists")

import xml.etree.ElementTree as ET
root = ET.parse(file_pth)

type(root)
```

```python
def find_child_nodes(node_name:str=None,parent=None):
    subnodes  = []
    if node_name is None:
        raise ValueError("name is null")

    
    if isinstance(parent,ET.ElementTree):
        return parent.findall(node_name)
    
    if isinstance(parent,ET.Element):
        return parent.findall(node_name)
    
    if isinstance(parent,list):
        for subnode in parent:
            subnodes.extend(subnode.findall(node_name))

    return subnodes
```

```python
titles = root.findall("title")
print("titles :\t",len(titles))

events  = find_child_nodes("event",titles)
print("events:\t",len(events))

relations = find_child_nodes("EventPair",titles)
print("relations:\t",len(relations))

# docs:	 350
# sentences:	 2502
# events:	 8327
# relations:	 6260
```

```python
def show_doc_statics(titles):
    for title in titles:
        events  = find_child_nodes("event",title)
        print("docs:\t",len(events),end='\t')

        relations = find_child_nodes("EventPair",title)
        print("relations:\t",len(relations))
    
    print("end .")

show_doc_statics(titles)
```
